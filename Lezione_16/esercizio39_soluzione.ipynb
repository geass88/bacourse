```python
# ALERT: execute this cell to prepare input data!
import requests
def download(link, nomeFile=None):
    if nomeFile == None:
        nomeFile = link.split('/')[-1]
    richiesta = requests.get(link)
    if richiesta.status_code == 200:
        with open(nomeFile, 'w') as file:
            file.write(richiesta.text)
            
download('https://tommasoadamo.it/data/grafo0.csv')
download('https://tommasoadamo.it/data/grafo1.csv')
download('https://tommasoadamo.it/data/grafo2.csv')
```

```python
import csv

def count_arcs(G):
    a = 0
    for r in G:
        for c in r:
            if c != 0:
                a+=1
    return a

def is_source(G, v):
    return in_d(G, v) == 0

def is_sink(G, v):
    return out_d(G, v) == 0

def in_n(G, v):
    return set(i for i in range(len(G)) if G[i][v]!=0)

def out_n(G, v):
    return set(i for i in range(len(G)) if G[v][i]!=0)

def in_d(G, v):
    return len(in_n(G, v))

def w_in_d(G, v):
    return sum(G[i][v] for i in in_n(G, v))

def w_out_d(G, v):
    return sum(G[v][i] for i in out_n(G, v))

def out_d(G, v):
    return len(out_n(G, v))

def top_bilanciato(G):
    for i in range(len(G)):
        if in_d(G, i) != out_d(G, i):
            return False
    return True

def bilanciato(G):
    for i in range(len(G)):
        if w_in_d(G, i) != w_out_d(G, i):
            return False
    return True

def è_un_cammino(G, cammino):
    for i in range(len(cammino)-1):
        if G[cammino[i]][cammino[i+1]] == 0:
            return False
    return True
    
def costo_cammino(G, cammino):
    costo = 0
    for i in range(len(cammino)-1):
        if G[cammino[i]][cammino[i+1]] != 0:
            costo += G[cammino[i]][cammino[i+1]]
        else:
            return None
    return costo

def leggi_grafo(nome):
    with open(nome, 'r') as file:
        reader = csv.reader(file, quoting=2)
        G = list(reader)
    return G

G=leggi_grafo('grafo0.csv')

#for i in range(len(G)):           # lo rendo sia topologicamente bilanciato che bilanciato 
#    for j in range(i, len(G)):
#        G[i][j] = G[j][i] 
        
print('Ci sono {} archi.'.format(count_arcs(G)))
for i in range(len(G)):
    print('Vertice {} in_degree={} in_neighbors={}, out_degree={} out_neighbors={}, is_sink={} is_source={}, weighted_in_degree={} weighted_out_degree={}'.format(i, in_d(G, i), in_n(G, i), out_d(G, i), out_n(G, i), is_sink(G, i), is_source(G, i),  w_in_d(G, i), w_out_d(G, i)))

if top_bilanciato(G):
    print('Il grafo è topologicamente bilanciato.')
else:
    print('Il grafo non è topologicamente bilanciato.')
    
if bilanciato(G):
    print('Il grafo è bilanciato.')
else:
    print('Il grafo non è bilanciato.')

cammino = [0, 1, 4, 5]
if è_un_cammino(G, cammino):
    print('{} è un cammino con costo {}'.format(cammino, costo_cammino(G, cammino)))
else:
    print('{} non è un cammino'.format(cammino))

cammino = [0, 1, 2, 3, 1, 6, 5]
if è_un_cammino(G, cammino):
    print('{} è un cammino con costo {}.'.format(cammino, costo_cammino(G, cammino)))
else:
    print('{} non è un cammino.'.format(cammino))              
```
